```python
import os
import json
import requests
import pandas as pd
from pathlib import Path

# 프로젝트 루트 기준 경로
ROOT = Path("../..").resolve()
RAW   = ROOT / "data" / "raw"
INTERIM = ROOT / "data" / "interim"
PROCESSED = ROOT / "data" / "processed"

for p in [RAW, INTERIM, PROCESSED]:
    p.mkdir(parents=True, exist_ok=True)

print("ROOT:", ROOT)
```

```python
# .env 파일에서 API 키 로드
# .env 형식:
#   DATAGOKR_NADAL_KEY=낱알식별키
#   DATAGOKR_EYAK_KEY=e약은요키

try:
    from dotenv import load_dotenv
    load_dotenv(ROOT / ".env")
except ImportError:
    pass

NADAL_KEY = os.environ.get("DATAGOKR_NADAL_KEY", "")
EYAK_KEY  = os.environ.get("DATAGOKR_EYAK_KEY", "")
print("낱알식별 키 설정됨:", bool(NADAL_KEY))
print("e약은요 키 설정됨: ", bool(EYAK_KEY))
```

```python
# openFDA 연결 확인 (인증 불필요)
resp = requests.get(
    "https://api.fda.gov/drug/label.json",
    params={"search": "openfda.generic_name:aspirin", "limit": 1},
    timeout=10
)
print("openFDA status:", resp.status_code)
if resp.ok:
    print("총 결과 수:", resp.json()["meta"]["results"]["total"])
```
